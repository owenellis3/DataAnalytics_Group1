```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,5,8,10,5]
y2 = [15,3,8,12,3]

if (len(x)==len(y)):
    plt.plot(x,y,label='ANA',color='mediumspringgreen')
    plt.plot(x,y2,label='BAL',color='y')
    plt.xlabel('Players')
    plt.ylabel('Height')
    plt.title('Baseball Player Heights')
    plt.legend()
    plt.show()
```

```python
import seaborn as sns
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

search_criteria = ['Position']
plotting_df = daf.find_aggregations(search_criteria)

ax = sns.barplot(y= plotting_df.index, x= plotting_df.max_height, orient = 'h', color='white')
ax.set_xlabel('Maximum Height')
ax.set_title('Maximum Height by Position')
ax.set_facecolor('k')
ax.set_xlim(65,85)
```

```python
plotting_df = daf.original_df
plotting_df.head(8)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax = sns.regplot(plotting_df.Height, plotting_df.Weight)
ax.set_title('Height v Weight')
```
